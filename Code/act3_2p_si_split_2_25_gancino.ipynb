# Activity 3: Data splitting - final model

## 1. Read Dataset and import libraries

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
```

```python
df_encoded = pd.read_csv('df_encoded.csv')
df_encoded
```

---------
Use the encoded dataset

```python
df_kfold = df_encoded.copy()

X = df_kfold.drop(columns=["rating"])
y = df_kfold["rating"]

X = np.array(X)
y = np.array(y)

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

---------
##2. KFold cross validation

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=42)

fold_number = 1
```

```python
for train_index, test_index in kfold.split(X):

    print(f"\n--- Fold {fold_number} ---")

    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    print("Train size:", len(X_train), " | Test size:", len(X_test))

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled  = scaler.transform(X_test)

    # Fit the model with the training data
    model = LinearRegression()
    model.fit(X_train_scaled, y_train)
    preds = model.predict(X_test_scaled)

    # Saving the metrics
    r2 = r2_score(y_test, preds)
    mse = mean_squared_error(y_test, preds)
    rmse = np.sqrt(mse)

    print("R²:", r2)
    print("MSE:", mse)
    print("RMSE:", rmse)

    fold_number += 1
```

##3. Leave-One-Out Cross Validation

```python
import warnings

warnings.filterwarnings("ignore", category=UserWarning, message=".*R² score is not well-defined.*")
```

Use de encoded dataset

```python
X = df_encoded.drop("rating", axis=1)
y = df_encoded["rating"]
```

---------
##4. Define the Leave-One-Out Cross Validation

```python
loo = LeaveOneOut()
```

Calling the model

```python
model_B = LinearRegression()

mse_scores = []
rmse_scores = []
```

---------
##5. Using Leave-One-Out Cross Validation

```python
predictions = []

for train_index, test_index in loo.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Fit the model with the training data
    model_B.fit(X_train, y_train)
    pred_B = model_B.predict(X_test)

    predictions.append(pred_B[0])

    # Saving the metrics
    mse_scores.append(mean_squared_error(y_test, pred_B))
    rmse_scores.append(np.sqrt(mean_squared_error(y_test, pred_B)))

# See the prediction
print("Predictions:")
for i, (pred, real) in enumerate(zip(predictions, y)):
    print(f"Sample {i+1} - Prediction: {pred}, Rating: {real}")
```

```python
print("Model B (Leave-One-Out Cross Validation):")
print("Average MSE:", np.mean(mse_scores))
print("Average RMSE:", np.mean(rmse_scores))

```
